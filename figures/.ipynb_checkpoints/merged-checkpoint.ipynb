```python
from collections import defaultdict
from collections import defaultdict
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter


from matplotlib.colors import LinearSegmentedColormap
import numpy as np
import pandas as pd
import seaborn as sns
import json
from scipy.cluster import hierarchy
import json
from collections import Counter

from scipy.spatial.distance import pdist
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import gzip
from collections import defaultdict
import pickle
import json
import seaborn as sns
import matplotlib.pyplot as plt
from glob import glob
import os
import numpy as np
import itertools
from tqdm import tqdm

import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import itertools
from scipy.stats import mannwhitneyu
from brokenaxes import brokenaxes

def config_params(font_size=7):

    mpl.rcParams.update(mpl.rcParamsDefault)
    plt.rcParams['font.sans-serif'] = ['arial']
    plt.rcParams['font.size'] = font_size
    plt.rcParams['font.family'] = ['sans-serif']
    plt.rcParams['svg.fonttype'] = 'none'
    plt.rcParams['mathtext.fontset'] = 'custom'
    plt.rcParams['mathtext.cal'] = 'arial'
    plt.rcParams['mathtext.rm'] = 'arial'
    
def get_age_patients(df):

    
    try:
        bio = df['biopsyDate'].split('-')[0]
        birth = df['birthYear']
        age = int(bio)-int(birth)
    except:
        age = 'unknown'
    
    return age

def return_age_treatment_dict(path_metadata):

    pat = pd.read_csv(path_metadata, sep='\t', encoding='latin-1')

    pat['primaryTumorLocation'] = pat['primaryTumorLocation'].replace('Bone/soft tissue', 'Bone/Soft tissue')

    # fix primary location
    pat['primaryTumorLocation_fixed'] = pat['primaryTumorLocation'].apply(
        lambda x: str(x).replace(' ', '-').replace('/', '-')
    )
    pat['primaryTumorLocation_fixed'] = pat['primaryTumorLocation_fixed'].replace('Head-and-Neck', 'Head-and-neck')
    pat['primaryTumorLocation_fixed'] = pat['primaryTumorLocation_fixed'].replace('nan', 'Unknown')
    pat['primaryTumorLocation_fixed'] = pat['primaryTumorLocation_fixed'].replace('CUP', 'Unknown')
    no_pretreat = pat[(pat['hasSystemicPreTreatment']=='No')&(pat['hasRadiotherapyPreTreatment']=='No')]
    no_pretreat_sample = no_pretreat['sampleId'].tolist()
    pretreat_sample = pat[(pat['hasSystemicPreTreatment']=='Yes')|(pat['hasRadiotherapyPreTreatment']=='Yes')]['sampleId'].tolist()

    pat['age'] = pat.apply(get_age_patients, axis = 1)
    
    dic_aging = dict(zip(pat['sampleId'], pat['age']))
    dic_gender = dict(zip(pat['sampleId'], pat['gender']))

    return no_pretreat_sample, pretreat_sample, dic_aging, dic_gender


def metadata_TCGA(clinical_data_tcga):
    
    df_t = pd.read_csv(clinical_data_tcga, sep ='\t', encoding = "ISO-8859-1", 
                      low_memory=False )

    aged = df_t[df_t['age_at_initial_pathologic_diagnosis']!='[Not Available]']
    aged = aged[aged['gender']!='[Not Available]']
    aged['age_at_initial_pathologic_diagnosis'] = aged['age_at_initial_pathologic_diagnosis'].astype(int)
    dic_aging = dict(zip(aged['bcr_patient_barcode'], aged['age_at_initial_pathologic_diagnosis']))
    dic_tobacco = dict(zip(aged['bcr_patient_barcode'], aged['tobacco_smoking_history']))

    mean_val = np.mean(list(dic_aging.values()))
    ttype_patients = dict(zip(df_t['bcr_patient_barcode'], df_t['acronym']))

    wanted_smoker = ['Current reformed smoker for < or = 15 years', 'Current smoker',
                    'Current reformed smoker for > 15 years', 'Current Reformed Smoker, Duration Not Specified']

    list_smoker = df_t[df_t['tobacco_smoking_history'].isin(wanted_smoker)]['bcr_patient_barcode'].tolist()
    nonsmoker = df_t[df_t['tobacco_smoking_history']=='Lifelong Non-smoker']['bcr_patient_barcode'].tolist()
    gender_patients = dict(zip(aged['bcr_patient_barcode'], aged['gender']))

    number_packs = df_t[(df_t['number_pack_years_smoked']!='[Not Available]')&
                        (~df_t['number_pack_years_smoked'].isnull())]
    smoking_history_years = dict(zip(number_packs['bcr_patient_barcode'],
                                     number_packs['number_pack_years_smoked']))
    
    return dic_aging, dic_tobacco, gender_patients

def classic_mutual_exclusivity_visualization(regression_table, sorted_treatments):

    df = regression_table.copy()
    df = df[sorted_treatments]
    X = df.values
    
    # for each gene give weights proportional to the number of mutations there
    weights = [1 / (i + 1) for i, c in enumerate(df.columns)]
    Y = pdist(X, metric='hamming', w=np.array(weights))
    linkage = hierarchy.linkage(Y, method='ward')
    g = sns.clustermap(df.T, col_linkage=linkage, col_cluster=True, 
                       row_cluster=False, figsize=(4, 6),  #col_colors = colors,
                       cmap = 'Blues', xticklabels = True, yticklabels = True )
    plt.close()

    return g
```

```python
from sklearn.metrics import jaccard_score

dic_res = defaultdict(dict)
two_samp = all_driv[all_driv['SAMPLE'].isin(shared_samples)]

for gene, data in two_samp.groupby(by='SYMBOL'):
    for gene2, data2 in two_samp.groupby(by='SYMBOL'):
        if (len(data2)>=5)&(len(data)>=5):
            s1_l = []
            s2_l = []
            if gene2 != gene:
                s1 = data['SAMPLE'].tolist() 
                s2 = data2['SAMPLE'].tolist()
                all_s = set(s1+s2)
                for s in all_s:
                    s1_v = 0
                    s2_v = 0
                    if s in s1:
                        s1_v = 1
                    if s in s2:
                        s2_v = 1
                    s1_l.append(s1_v)
                    s2_l.append(s2_v)

                score = jaccard_score(s1_l, s2_l)
                dic_res[gene][gene2] = score

res_jac = pd.DataFrame(dic_res)
res_jac = res_jac.loc[res_jac.columns]
```

```python
def process_discovery(df_subs, name):
    
    #df = pd.read_csv(file, sep ='\t')
    
    df = df_subs[['CHROM', 'POS', 'REF', 'ALT', 'SAMPLE']]
    # select the max frequency we would expect (in this case, the DNMT3A hotspot)
    len_DNMT3A = len(df[(df['CHROM'] =="2") & (df['POS']==25457242)])
    df['ID'] = df.apply(lambda x: '{}_{}_{}'.format(x['CHROM'], x['POS'], x['ALT']), axis = 1)
    dic_counts = df['ID'].value_counts().to_dict()
    forbidden_hotspots = [k for k, v in dic_counts.items() if v>len_DNMT3A]
    df = df[~df['ID'].isin(forbidden_hotspots)]
        
    df['CHROMS'] = df['CHROM'].apply(lambda x : 'chr{}'.format(x))
    df['S'] = df['POS'] -1 
    df.columns = ['chromo', 'end', 'ref', 'alt', 'id', 'n',
            '#chr','start']
    df[['#chr', 'start', 'end', 'ref', 'alt', 'id']].to_csv('{0}/DriverPower/{1}.tsv'.format(path_in, name), 
                                                                        sep ='\t', index = False, header = True )

    df.columns = ['chr', 'pos', 'ref_allele', 'newbase', 'patient', 'n',
            '#chr','start']
    df[['chr', 'pos', 'patient', 'ref_allele', 'newbase']].to_csv('{0}/MutSig2CV_NC/{1}.tsv'.format(path_in,name), 
                                                                        sep ='\t', index = False, header = True )

    df.columns = ['#chr', 'pos2', 'ref', 'alt', 'patient', 'n',
            'chr','pos1']
    df[['chr', 'pos1', 'pos2', 'ref', 'alt',  'patient']].to_csv('{0}/ActiveDriverWGS/{1}.tsv'.format(path_in, name), 
                                                                        sep ='\t', index = False, header = True )
    
    df.columns = ['CHROMOSOME', 'POS', 'REF', 'ALT', 'SAMPLE', 'n',
            'chr','pos1']
    df['CHROMOSOME'] = df['CHROMOSOME'].apply(lambda x : str(x).replace('chr', ''))
    df[['CHROMOSOME', 'POS', 'REF', 'ALT', 'SAMPLE']].to_csv('{0}/oncodriveFML/{1}.tsv'.format(path_in, name), 
                                                                        sep ='\t', index = False, header = True )
       
    
```

```python
import pandas as pd
from statsmodels.stats.multitest import multipletests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import fisher_exact
import matplotlib as mpl
import matplotlib.pyplot as plt

# config for matplotlib
def config_params(font_size=7):
    mpl.rcParams.update(mpl.rcParamsDefault)
    plt.rcParams['font.sans-serif'] = ['arial']
    plt.rcParams['font.size'] = font_size
    plt.rcParams['font.family'] = ['sans-serif']
    plt.rcParams['svg.fonttype'] = 'none'
    plt.rcParams['mathtext.fontset'] = 'custom'
    plt.rcParams['mathtext.cal'] = 'arial'
    plt.rcParams['mathtext.rm'] = 'arial'
    
    
config_params(font_size=5.5)
```

```python
def eliminate_duplicates(df):

    colors = df['color'].tolist()
    x = df['x'].tolist()[0]
    y = df['y'].tolist()[0]
    if 'r' in colors:
        return x, y, 'red'
    elif 'g' in colors:
        return x, y, 'green'
    else:
        return x, y, 'grey'
    
def plot_qqplot(df, pval, qval, gene, name, many):
    
    pvalue=pval
    qvalue=qval
    min_samples = 2
    draw_greys = True
    annotate = True
    cut = True
    #############################

    MIN_PVALUE = 10000
    MAX_ANNOTATED_GENES = 50

    df.dropna(subset=[pvalue], inplace=True)    
    
    res = []
    valsFisher = []
    
    if len(df)>0:
        
        # genes significant
        res = list(set(df[df[qval]<=0.15][gene].tolist()))

        # Define the shape of the figure
        NROW = 1
        NCOL = 1

        fig = plt.figure(figsize=(2.75, 2.27))
        axs = [plt.subplot2grid((NROW, NCOL), (item // NCOL, item % NCOL)) for item in range(NROW * NCOL)]
        
        plt.title(name)
        # Plot is on the right or on the left?
        dx_side = True
        ax = axs[0]

        colors = ['grey', 'black']
        obs_pvalues = df[pvalue].map(lambda x: -np.log10(x) if x > 0 else -np.log10(1 / MIN_PVALUE))
        #gens with num_samples >= min_samples have different values for color and alpha
        obs_color = colors[1] # df['SAMPLES'].map(lambda x: colors[1] if x >= min_samples else colors[0])
        obs_alpha = 0.7  #df['SAMPLES'].map(lambda x: 0.7 if x >= min_samples else 0.3)

        data = pd.DataFrame({'HugoID': df[gene],
                             'observed': obs_pvalues,
                             'color': obs_color,
                             'alpha': obs_alpha,
                             'fdr': df[qvalue] if qvalue is not None else 1
                             })

        data.sort_values(by=['observed'], inplace=True)
        exp_pvalues = -np.log10(np.arange(1, len(data) + 1) / float(len(data)))
        exp_pvalues.sort()
        data['expected'] = exp_pvalues

        # Get the maximum pvalues (observed and expected)
        max_x = float(data[['expected']].apply(np.max))
        max_y = float(data[['observed']].apply(np.max))

        # Give some extra space (+-5%)
        min_x = max_x * -0.05
        min_y = max_y * -0.05
        max_x *= 1.1
        max_y *= 1.1

        grey = data[data['color'] == colors[0]]
        blue = data[data['color'] == colors[1]]

        # Plot the data
        if draw_greys and len(grey['expected']) > 0:
            ax.scatter(grey['expected'].tolist(),
                       grey['observed'].tolist(),
                       color=grey['color'].tolist(),
                       alpha=grey['alpha'].tolist()[0],
                       s=12)

        if len(blue['expected']) > 0:
            ax.scatter(blue['expected'].tolist(),
                       blue['observed'].tolist(),
                       color=blue['color'].tolist(),
                       alpha=blue['alpha'].tolist()[0],
                       s=12)

        # Get the data that are significant with a FDR < 0.1 and FDR 0.25
        genes_to_annotate = []
        for fdr_cutoff, fdr_color in zip((0.25, 0.2), ('g-', 'r-')):
            fdr = data[data['fdr'] < fdr_cutoff]['observed']
            if len(fdr) > 0:
                fdr_y = np.min(fdr)
                fdr_x = np.min(data[data['observed'] == fdr_y]['expected'])
                ax.plot((fdr_x - max_x * 0.025, fdr_x + max_x * 0.025), (fdr_y, fdr_y), fdr_color)
                # Add the name of the significant genes
                genes = data[(data['observed'] >= fdr_y) & (data['expected'] >= fdr_x)]
                for count, line in genes.iterrows():
                    if line['color'] == colors[0]:
                        continue
                    genes_to_annotate.append({'x': line['expected'],
                                              'y': line['observed'],
                                              'HugoID': line['HugoID'],
                                              'color': colors[0] if line['color'] == colors[0] else fdr_color[0]})

        # Annotate the genes
        genes_to_annotate = genes_to_annotate[-many:]
        genes_annotated = 0
        if annotate and len(genes_to_annotate) > 0:
            genes_to_annotate = pd.DataFrame(genes_to_annotate)

            # Get rid of redundant genes
            grouped = genes_to_annotate.groupby(['HugoID'])
            grouped = grouped.apply(eliminate_duplicates)
            grouped = pd.DataFrame({'HugoID': grouped.index.tolist(),
                                    'x': [x[0] for x in grouped],
                                    'y': [y[1] for y in grouped],
                                    'color': [c[2] for c in grouped]})
            grouped.sort_values(by=['y', 'x'], inplace=True, ascending=[False, False])

            x_text = max_x * 1.1 if dx_side is True else min_x - (max_x * 0.2)
            y_text = np.floor(max_y)
            distance_between_genes = max_y * 0.05
            for count, line in grouped.iterrows():
                x, y = line['x'], line['y']
                ax.annotate(line['HugoID'], xy=(x, y), xytext=(x_text, y_text),
                            arrowprops=dict(facecolor="black", headlength = 0.02, shrink=0.1,
                                            width=0.3, headwidth=0, alpha=0.3),
                            horizontalalignment="left", verticalalignment="center",

                            color=line['color'],
                            weight = 'normal', 
                            fontsize = 6
                            )
                y_text -= distance_between_genes

                # This avoid the crash for ValueError: width and height must each be below 32768
                genes_annotated += 1
                if genes_annotated >= MAX_ANNOTATED_GENES:
                    logger.warning("Annotations cut to %d genes", MAX_ANNOTATED_GENES)
                    break

        # Add labels
        ax.set_xlabel("-log10(pvalues) expected")
        ax.set_ylabel("-log10(pvalues) observed")

        # Add the dashed diagonal
        ax.plot(np.linspace(0, np.floor(max(max_x, max_y))),
                np.linspace(0, np.floor(max(max_x, max_y))),
                'darkred', ls ='--')
        ax.grid(True)

        # Redefine the limits of the plot
        #if cut:
        #    ax.set_xlim(min_x, max_x)
        #    ax.set_ylim(min_y, max_y)

        ax.set_xlim(min_x, max_x)
        ax.set_ylim(min_y, max_y)
        # Set the title: project, cancer_type
        #ax.set_title(title)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

        # Adjust the plot
        plt.tight_layout()
        plt.show()
```

```python
files = glob('/workspace/datasets/reverse_calling/noncoding/software/MutSig2CV_NC/HMF_not_repetitive_mosaic/Hematopoietic_tumors_*/results.verbose.mat')

toconcat = []
for f in tqdm(files):
    if '_TFBS' not in f:
        annots = loadmat(f)
        header = [('name', 'O'), ('chr', 'O'), ('start', 'O'), ('end', 'O'), ('exon_starts', 'O'),
      ('exon_ends', 'O'), ('rt_chen_extra1_max', 'O'), ('rt_amnon_max', 'O'), ('hiC', 'O'), ('exprsum', 'O'), ('exprmax', 'O'), ('exprmed', 'O'), ('gc', 'O'), ('paz', 'O'), ('rt', 'O'), ('log_exprsum', 'O'), ('log_exprmax', 'O'), ('log_exprmed', 'O'), ('nstrikes', 'O'), ('min_effect_size', 'O'), ('codelen', 'O'), ('Nssnv', 'O'), ('Nind', 'O'), ('nssnv', 'O'), ('nind', 'O'), ('ntest', 'O'), ('Ntest', 'O'), ('nbagel', 'O'), ('Nbagel', 'O'), ('nnei', 'O'), ('nnei2', 'O'), ('npat_exp', 'O'), ('npat', 'O'), ('nsite', 'O'), ('freq', 'O'), ('freq_excess', 'O'), ('x', 'O'), ('X', 'O'), ('Fg_post', 'O'), ('S', 'O'), ('pCVmin', 'O'), ('pCVmid', 'O'), ('pCVmax', 'O'), ('nmut', 'O'), ('sCL', 'O'), ('sFN', 'O'), ('pFN_Gaussian_m', 'O'), ('pFN_Gaussian_s', 'O'), ('pFN_ocons', 'O'), ('pCV_Gaussian', 'O'), ('pCV_Gaussian_m', 'O'), ('pCV_Gaussian_s', 'O'), ('pCV_score_obs', 'O'), ('pCV_mefsz', 'O'), ('nperm', 'O'), ('pCL', 'O'), ('pFN', 'O'), ('pCLFN', 'O'), ('pCL2', 'O'), ('pFN2', 'O'), ('pCLFN2', 'O'), ('pCF', 'O'), ('shortname', 'O'), ('pFN3', 'O'), ('pmid', 'O'), ('pmax', 'O'), ('qCF', 'O'), ('qCV', 'O'), ('qCL', 'O'), ('qFN', 'O'), ('qCL2', 'O'), ('qFN2', 'O'), ('q', 'O'), ('nnon', 'O')]

        rel_header = [x[0] for x in header]

        toappend = []
        good_el = []
        real_ids = 0
        for ix, l in enumerate(annots['G'][0][0]):
            if ix == 0:
                real_ids = len(l)
            try:
                flat_list = [item[0] for sublist in l for item in sublist]
            except:
                flat_list = [item for sublist in l for item in sublist]

            if len(flat_list)==real_ids:
                toappend.append(flat_list)
                good_el.append(ix)

        out = pd.DataFrame(toappend).T
        head = [rel_header[ix] for ix in good_el]
        out.columns = head
        toconcat.append(out)
        
merged = pd.concat(toconcat)
merged['TYPE_ELEMENT'] = merged['name'].apply(lambda x : x.split(':')[0])
out_mutect = merged[merged['TYPE_ELEMENT'].isin(wanted)]
res = multipletests(out_mutect['pCVmax'].tolist(), method ='fdr_bh')
out_mutect['qtesting'] = res[1]
out_mutect['NAME_TO_PLOT'] = out_mutect['name'].apply(lambda x : '{}:{}'.format( x.split('::')[2], 
                                                  x.split('::')[0].replace('gc19_pc.', '' )) if 'enhancers' not in x 
                                                        else '{}:{}'.format( x.split('::')[1], 
                                                  x.split('::')[0]
                                                        ))


```

# Merge results

```python
drivep = pd.read_csv('/home/opich/Nextcloud/AML/split/CH/noncoding/discovery/driverpower_mosaic.tsv', sep ='\t')
mutsig = pd.read_csv('/home/opich/Nextcloud/AML/split/CH/noncoding/discovery/mutsigcv_mosaic.tsv', sep ='\t')
fml = pd.read_csv('/home/opich/Nextcloud/AML/split/CH/noncoding/discovery/oncodrivefml_mosaic.tsv', sep ='\t')

```

```python
from collections import defaultdict
from pybedtools import BedTool
```

```python
f = BedTool('/workspace/datasets/reverse_calling/noncoding/software/DriverPower/HMF_not_repetitive_mosaic.tsv')
location_genes = '/workspace/datasets/reverse_calling/noncoding/software/DriverPower/test_elements.tsv'

df_loc = pd.read_csv(location_genes, sep ='\t', names = ['chr', 'start', 'end', 
                                                        'ID'])
df_loc['ELEMENT'] = df_loc['ID'].apply(lambda x : x.split('::')[0])
df_loc['LEN'] = df_loc['end']- df_loc['start']
```

```python
df_loc['coord'] = df_loc.apply(lambda x : '{}:{}-{}'.format(x['chr'], x['start'], x['end']), 
                              axis = 1)
```

```python
fml_genes = fml['GENE_ID'].tolist()
mutsig_genes = mutsig['name'].tolist()
drivep_list = drivep['binID'].tolist()
dic_res = defaultdict(dict)
all_driv = set(fml_genes + mutsig_genes + drivep_list)
```

```python
dic_coord = {}
for element, data in df_loc[df_loc['ID'].isin(all_driv)].groupby(by='ID'):
    l = ';'.join(data['coord'].tolist())
    dic_coord[element] = l
    
```

```python
df_loc_mutect = BedTool.from_dataframe(df_loc[df_loc['ID'].isin(all_driv)])
out =df_loc_mutect.intersect(f, wo = True)
intersect = out.to_dataframe(header = None)
intersect['ELEMENT'] = intersect[3].apply(lambda x : x.split('::')[2])
nmuts_element = intersect[3].value_counts().to_dict()
```

```python

for g in all_driv:
    dic_res[g]['OncodriveFML_q'] = dic_fml_q[g]

    try:
        dic_res[g]['DriverPower_q'] = dic_drivepower_q[g]
        dic_res[g]['MutSigCV_q'] = dic_mutsig_q[g]

    except:
        dic_res[g]['DriverPower_q'] = dic_drivepower_q[g.replace('enhancers::', 'enhancers::chr')]
        dic_res[g]['MutSigCV_q'] = dic_mutsig_q.get(g.replace('enhancers::', 'enhancers::chr'), 1)

    dic_res[g]['Element_ID'] = g
    dic_res[g]['Number_mutations'] = nmuts_element[g]
    dic_res[g]['Coordinates'] = dic_coord[g]

```

```python
result = pd.DataFrame(dic_res).T

result['SYMBOL'] = result['Element_ID'].apply(lambda x : '{}'.format( x.split('::')[2]) 
                                                  if 'enhancers' not in x 
                                                        else '{}'.format( x.split('::')[1] 
                                                        ))

result['Type_Element'] = result['Element_ID'].apply(lambda x : x.split(':')[0])
```

```python
result = result.sort_values(by=['OncodriveFML_q', 'MutSigCV_q', 'DriverPower_q'])
```

```python
final_res = result[['SYMBOL', 'Type_Element', 'Element_ID', 'OncodriveFML_q', 'MutSigCV_q', 'DriverPower_q', 'Number_mutations', 'Coordinates']]
```

```python
final_res.to_csv('/home/opich/Nextcloud/AML/to_biorxiv/Tables/supplementary_table_3.tsv', 
                sep ='\t', header = True, index = False)
```
