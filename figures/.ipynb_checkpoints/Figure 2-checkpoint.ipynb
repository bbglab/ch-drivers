```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import json

# config for matplotlib
def config_params(font_size=7):
    mpl.rcParams.update(mpl.rcParamsDefault)
    plt.rcParams['font.sans-serif'] = ['arial']
    plt.rcParams['font.size'] = font_size
    plt.rcParams['font.family'] = ['sans-serif']
    plt.rcParams['svg.fonttype'] = 'none'
    plt.rcParams['mathtext.fontset'] = 'custom'
    plt.rcParams['mathtext.cal'] = 'arial'
    plt.rcParams['mathtext.rm'] = 'arial'
    
```

```python
# files obtained in the process_data module
file_reverse_calling_TCGA = '/Users/picho/projects/collaborations/bbglab/CH/clonal_hematopoiesis_code/data_VAF/TCGA/vaf_filter/TCGA_FULL/TCGA_full.vep.sorted.tsv.filtered.toUSE.gz'
file_reverse_calling_HMF = '/Users/picho/projects/collaborations/bbglab/CH/clonal_hematopoiesis_code/data_VAF/HMF/vaf_filter/HMF_FULL/HMF_full.vep.sorted.tsv.filtered.toUSE.gz'

df1 = pd.read_csv(file_reverse_calling_TCGA, sep ='\t')

df1['COHORT'] = 'TCGA'
df2 = pd.read_csv(file_reverse_calling_HMF, sep ='\t')
df2['COHORT'] = 'HMF'

df_muts = pd.concat([df1, df2])

# mutations from cbioportal
mskcc = pd.read_csv('~/projects/collaborations/bbglab/CH/mutpro/data/other_CH_datasets/data_mutations_extended.txt',
                    skiprows = 2, sep ='\t')
mskcc = mskcc[~mskcc['Consequence'].isin(['3_prime_UTR_variant', 'downstream_gene_variant','upstream_gene_variant', 
                               '5_prime_UTR_variant', 'intron_variant', 'inframe_insertion', 
                               'inframe_deletion', 'synonymous_variant'])]

dic_CH_genes = json.load(open('../data/tables/list_genes_CH_category.json'))
final_res = pd.read_csv("../data/tables/supplementary_table_2.tsv", sep ="\t")
pass_filter = pd.read_csv("../data/tables/driver_tier1.tsv", sep ='\t')
```

```python
dic_res = defaultdict(dict)
all_genes = final_res['Gene'].unique()
for g in all_genes:
    
    dic_res[g]['Primary'] = 0
    dic_res[g]['Metastasis'] = 0
    dic_res[g]['Targeted'] = 0
    dic_res[g]['CH'] = 0
    dic_res[g]['Myeloid'] = 0
    dic_res[g]['CGC'] = 0
    
    if len(pass_filter[(pass_filter['SYMBOL']==g)&(pass_filter['COHORT'].str.contains('HMF'))])>0:
        dic_res[g]['Metastasis'] = 2
    else:
        if len(df_muts[(df_muts['SYMBOL']==g)&(df_muts['COHORT']=='HMF')])>2:
            dic_res[g]['Metastasis'] = 1
            
    if len(pass_filter[(pass_filter['SYMBOL']==g)&(pass_filter['COHORT'].str.contains('TCGA'))])>0:
        dic_res[g]['Primary'] = 2
    else:
        if len(df_muts[(df_muts['SYMBOL']==g)&(df_muts['COHORT']=='TCGA')])>2:
            dic_res[g]['Primary'] = 1  
    if len(final_res[(final_res['Gene']==g)&(final_res['Discovery_IMPACT']=='YES')])>0:
        dic_res[g]['Targeted'] = 2
    else:
        if len(mskcc[(mskcc['Hugo_Symbol']==g)])>0:
            dic_res[g]['Targeted'] = 1
            
    if g in dic_CH_genes['CH']:
        dic_res[g]['CH'] = 1
    if g in dic_CH_genes['Myeloid']:
        dic_res[g]['Myeloid'] = 1    
    if g in dic_CH_genes['CGC']:
        dic_res[g]['CGC'] = 1

res = pd.DataFrame(dic_res).T
res['Evidence'] = res['Primary'] + res['Metastasis'] + res['Targeted']
res = res.sort_values(by=['CH', 'Myeloid', 'CGC', 'Evidence', ], ascending = [False, False, False, False])
res.drop('Evidence', axis = 1, inplace = True)
res = res.T
```

```python
config_params(font_size=7)

fig, ax = plt.subplots(1, 1, figsize = (10, 1))
prim = res.loc['Primary']
meta = res.loc['Metastasis']
target = res.loc['Targeted']
ch = res.loc['CH']
myel = res.loc['Myeloid']
cgc = res.loc['CGC']
  
color_signals_met = {0:'white', 1:'#E2E2E2', 2:'#E1AE06'}
color_signals_prim = {0:'white', 1:'#E2E2E2', 2:'#67B0E9'}
color_signals_targ = {0:'white',1:'#E2E2E2', 2:'#AA3114'}

color_cgc = {0:'white', 1:'#0E1137'}
color_myeloid = {0:'white', 1:'#C83737'}
color_ch = {0:'white', 1:'#548F25'}

for i in range(len(res.T)):
    if (i % 2)==0:
        plt.vlines(x = i, ymin = 0.55, ymax= 6.5, color = 'grey', alpha = 0.25, zorder = 0, 
                  linewidth = 8)
        
for ix, v in enumerate(res.columns.tolist()):
    plt.scatter(ix,6,  color = color_signals_prim[prim[v]], edgecolors='black', lw = 0.25)
    plt.scatter(ix,5, color = color_signals_met[meta[v]], edgecolors='black', lw = 0.25)
    plt.scatter(ix,4, color = color_signals_targ[target[v]], edgecolors='black', lw = 0.25)

    plt.scatter(ix,3,color = color_cgc[ch[v]], edgecolors='black', lw = 0.25)
    plt.scatter( ix,2,color = color_myeloid[myel[v]], edgecolors='black', lw = 0.25)
    plt.scatter(ix,1,color = color_ch[cgc[v]], edgecolors='black', lw = 0.25)

plt.xlim(-1, len(res.T))
plt.xticks(np.arange(len(res.T)),["$\it{0}$".format(lab) for lab in res.columns.tolist()], 
          rotation = 90)
plt.yticks([i +1 for i in np.arange(len(res))], 
           [ 'CGC','Myeloid','CH','Targeted','Metastasis','Primary'],
           )
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
    
```
